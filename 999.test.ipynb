```python

def get_all_file_paths(directory):
    # directory 경로의 모든 파일들의 경로 리스트 반환
    file_paths = []
    
    for root, dirs, files in os.walk(directory):
        # 현재 경로에 있는 모든 파일의 경로를 추가
        for file in files:
            file_paths.append(os.path.join(root, file))
            
        # 하위 디렉토리로 더 이상 내려가지 않기 위해서 break 사용
        break
    
    return file_paths

# output_dir 경로 설정
output_dir = '/root/lip-to-speech/split_TS163/lip_K_5_F_06_C651_A_001.mp4'

# 경로 리스트 가져오기
file_paths = get_all_file_paths(output_dir)
print(file_paths)
```
